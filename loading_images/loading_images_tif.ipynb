```python
# Imports
from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append(r'C:\Program Files\ASAP 1.9\bin')
import multiresolutionimageinterface as mir
```

```python
def open_multiresolutionimage_image(path: Path) -> mir.MultiResolutionImage:
    """Opens a multiresolution image with ASAP python bindings

    Args:
        path (Path): path to image

    Raises:
        IOError: raises when opened image is None

    Returns:
        MultiEesolutionImage: opened multiresolution image
    """
    reader = mir.MultiResolutionImageReader()
    image = reader.open(str(path))
    if image is None:
        raise IOError(f"Error opening image: {path}, image is None")
    return image
```

```python
# <Adapt this path>
data_path = r'./data/wsitils/images/104S.tif'

img = open_multiresolutionimage_image(data_path)
```

```python
# Get the width and the height of the image
x, y = img.getDimensions()

# Convert a patch of the image to an numpy array
image_tile = img.getUCharPatch(startX=0,
                               startY=0,
                               width=1000,     # Max y width of the patch
                               height=1000,    # Max x width of the patch
                               level=4         # Some obscure zoom
                              )
print(f'The type of the tile is: {type(image_tile)}')
```
